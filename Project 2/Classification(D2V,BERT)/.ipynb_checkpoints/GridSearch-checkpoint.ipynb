# Wisconsin Uinversity- Breast Cancer

```python
import pandas as pd 
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import statistics as st
```

```python
names=['ID', 'Clump Thickness', 'Uniformity of Cell Size', 'Uniformity of Cell Shape', 'Marginal Adhesion', 'Single Epithelial Cell Size'
      ,'Bare Nuclei', "Bland Chromatin",'Normal Nucleoli', 'Mitoses', 'Class']
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data", names=names)
```

```python
df.head()
```

```python
df.set_index("ID",inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Bare Nuclei'].unique()
```

```python
df['Bare Nuclei'][df['Bare Nuclei']=="?"]=st.mode(df['Bare Nuclei'])
```

```python
df['Bare Nuclei']=df['Bare Nuclei'].astype(np.int)
```

```python
df.info()
```

```python
df['Class'][df['Class']==2]=0
df['Class'][df['Class']==4]=1
df['Class']=[bool(i) for i in df['Class']]
```

```python
df.head()
```

```python
df.info()
```

```python
df.groupby('Class').count()
```

### Note: In the data set, Class 'False' means Benign tumor while 'True' means Malignant Tumor

```python
df.to_csv('Wis-breast_cancer.data')
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test=train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.2,random_state=14231)
```

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rf=RandomForestClassifier()
n=[i+1 for i in range(25)]
mfeat=['auto', 'log2', None]
param={'n_estimators':n, "max_features": mfeat}
```

```python
rf_grid=GridSearchCV(estimator=rf,param_grid=param, scoring='accuracy', n_jobs=-1, verbose=3,cv=5)
```

```python
rf_grid.fit(X_train,y_train)
```

```python
rf_grid.best_params_
```

```python
predictions=rf_grid.predict(X_test)
```

```python
print(metrics.classification_report(y_test,predictions))
```

```python
print('The accuracy Score is ' + str(metrics.accuracy_score(y_test,predictions)))
```

```python
help(rf_grid)
```

```python
results=pd.DataFrame(rf_grid.cv_results_)
```

```python
results
```

```python

plt.plot([i for i in range(X_test.shape[0])],y_test, 'ro', label='True')
plt.plot([i for i in range(X_test.shape[0])],predictions, 'bo', label='Predicted')
plt.legend()
plt.ylabel('Benign/Malignant')
plt.title('The performance of the model on Test Dataset')
plt.show()
```
